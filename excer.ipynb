```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.bbc.com/news"
```

```python
response = requests.get(url)
```

```python
raw_code = response.text
```

```python
soup =BeautifulSoup(raw_code,'lxml')
```

```python
print(soup.title)
```

```python
get_h2 = soup.find('h2').get_text()
```

```python
print(get_h2)
```

```python
var_2 = soup.find_all(class_="section")
```

```python
print(var_2)
```
